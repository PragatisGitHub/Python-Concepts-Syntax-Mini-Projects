```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Load in csv
path = "../Resources/avg_rain_state.csv"
AvgRain = pd.read_csv(path)
AvgRain.head()
```

```python
x_axis = np.arange(len(AvgRain["State"]))
```

```python
x_ticks = [x+0.4 for x in x_axis]
```

```python
y_axis = AvgRain["Inches"]
```

```python
plt.figure(figsize=(20,3))
plt.bar(x_axis,y_axis,color="red",alpha=0.5,align="edge")
```

```python
plt.title("Average Rain Per State")
plt.xlabel("State")
plt.ylabel("Average amount of rain in inches")
plt.xticks(x_ticks,AvgRain["State"],rotation="vertical")
```

```python
# Set x and y limits
plt.xlim(-1.25,len(x_axis)+1.25)
plt.ylim(0,max(AvgRain["Inches"]+10))
```

```python
plt.tight_layout()
```
