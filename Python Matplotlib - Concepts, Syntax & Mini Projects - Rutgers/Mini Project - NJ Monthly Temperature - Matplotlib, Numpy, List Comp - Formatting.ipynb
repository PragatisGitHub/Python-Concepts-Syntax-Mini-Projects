```python
import matplotlib.pyplot as plt

import numpy as np

# %matplotlib notebook
```

```python
plt.style.use('fivethirtyeight')
```

```python
Month = list(np.arange(1,13,1))
```

```python
Month
```

```python
Fahrenheit = [39, 42, 51, 62, 72, 82, 86, 84, 77, 65, 55, 44]
```

```python
Fahrenheit
```

```python
Celcius = [round((Fah - 39) * 0.56,2) for Fah in Fahrenheit]
```

```python
Celcius
```

```python
# %matplotlib notebook
```

```python
plt.xticks(Month)
```

```python
%matplotlib notebook
```

```python
Fahrenheit_plot,=plt.plot(Month,Fahrenheit,linewidth=2,label='Fahrenheit',marker='*',markersize=20,color="Blue",alpha=.5)
```

```python
Celcius_plot,=plt.plot(Month,Celcius,linewidth=2,label='Celcius',marker='o',markersize=20,color="Red",alpha=.5)
```

```python
plt.title("Temperature Variation in F & C")
plt.xlabel("Months")
plt.ylabel("Temperature")
```

```python
plt.legend(handles=[Fahrenheit_plot,Celcius_plot],loc='best')
```
