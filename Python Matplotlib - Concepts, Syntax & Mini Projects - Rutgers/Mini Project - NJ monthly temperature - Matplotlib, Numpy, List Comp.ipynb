```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
%matplotlib notebook
```

```python
Month = list(np.arange(1,13))
```

```python
Month
```

```python
Fahrenheit = [39, 42, 51, 62, 72, 82, 86, 84, 77, 65, 55, 44]
```

```python
plt.plot(Month,Fahrenheit)
```

```python
Celcius = [round((F - 39) * 0.56,2) for F in Fahrenheit]
```

```python
Celcius
```

```python
plt.plot(Month,Celcius)
```
