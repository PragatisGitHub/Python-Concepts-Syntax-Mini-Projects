```python
DangerDrop = [9, 8, 90, 85, 80, 70, 70, 65, 55, 60, 70, 65, 50]
type(DangerDrop)
max_DD = max(DangerDrop)
min_DD = min(DangerDrop)
```

```python
RailGun = [75, 70, 60, 65, 60, 45, 55, 50, 40, 40, 35, 35, 30]
type(RailGun)
max_RG = max(RailGun)
min_RG = min(RailGun)
```

```python
TimeMeasured = np.arange(0,130,10)

TimeMeasuredList = list(TimeMeasured)
max_time = max(TimeMeasuredList)
min_time = min(TimeMeasuredList)
```

```python
if min_DD < min_RG:
    ymin = min_DD
else:
    ymin = min_RG
if max_DD > max_RG:
    ymax = max_DD
else:
    ymax = max_RG    
print(max_time)
print(min_time)
print(ymin)
print(ymax)    
```

```python
# Plot the charts and apply some styling
DangerDrop,=plt.plot(TimeMeasuredList,DangerDrop,color="red",label="DangerDrop")
RailGun,=plt.plot(TimeMeasuredList,RailGun,color="green",label="RailGun")
```

```python
# Add labels to X and Y axes :: Add title
plt.title("Coaster Speed Over Time")
plt.xlabel("Coaster Runtime")
plt.ylabel("Speed (MPH)")
# Set the limits for the X and Y axes
plt.xlim(min_time,max_time)
plt.ylim(0,ymax+5)
# Add in a grid for the chart
plt.grid()
# Create a legend for the chart
plt.legend(handles=[DangerDrop,RailGun],loc="best")
plt.savefig("Roller Coaster.jpg")
```
