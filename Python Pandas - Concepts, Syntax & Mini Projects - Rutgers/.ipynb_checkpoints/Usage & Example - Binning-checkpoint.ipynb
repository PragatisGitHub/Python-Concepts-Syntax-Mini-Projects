```python
# Import Dependencies
import pandas as pd
```

```python
raw_data = {
    'Class': ['Oct', 'Oct', 'Jan', 'Jan', 'Oct', 'Jan','Feb','Mar','April','May','June','July'], 
    'Name': ["Cyndy", "Logan", "Laci", "Elmer", "Crystle", "Emmie","Soni","Mehta","Thakkar","Shah","Sheth","Parekh"], 
    'Test Score': [90, 56, 72, 88, 98, 67, 0, 60, 70, 80, 90, 100]}
df = pd.DataFrame(raw_data)
df
```

```python
# Create the bins in which Data will be held
# Bins are 0, 60, 70, 80, 90, 100
bins = [0, 60, 70, 80, 90, 100]

# Create the names for the four bins
group_names = ["F", "D", "C", "B", "A"]
```

```python
df["Test Score Summary"] = pd.cut(df["Test Score"], bins, labels=group_names)
df
```

```python
# Creating a group based off of the bins
df = df.groupby("Test Score Summary")
```
